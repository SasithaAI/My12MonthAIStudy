```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, skewnorm
```

```python
def get_data(n):
    data = np.concatenate((expon.rvs(scale=1, size=n//2), skewnorm.rvs(5, loc=3, size=n//2)))
    np.random.shuffle(data)
    return data
plt.hist(get_data(2000), bins=50);
```
